```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
#from sklearn.ensemble import RandomForestClassifier
#from sklearn.svm import SVC
#from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Co2_CO_data.csv')
df.head()
```

```python
df.insert(5,'Air_Quality',0)
df.head()
```

```python
co = df['CO_PPM']
co2 = df['Co2_PPM']
```

```python
for i in range(len(co2)):
  if (co[i] < 1.5 or co2[i] < 450):
    df.loc[i,'Air_Quality']= 0
  if (co[i] > 3 and co[i] < 5) or (co2[i] > 600 and co2[i] < 800):
    df.loc[i,'Air_Quality'] = 1
  if  (co[i] > 5 and co[i] < 8) or (co2[i] > 800 and co2[i] < 1200):
    df.loc[i,'Air_Quality'] = 2
  if  (co[i] > 8 and co[i] < 15) or (co2[i] > 1200 and co2[i] < 1500):
    df.loc[i,'Air_Quality'] = 3
  if (co[i] > 15):
    df.loc[i,'Air_Quality'] = 4
```

```python
df.head(50)
```

```python
X = df[['CO_PPM', 'Co2_PPM']]
y = df['Air_Quality']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.9, random_state=42)
```

```python
model = LogisticRegression(multi_class='multinomial', solver='newton-cg', C=1.0)
#model = RandomForestClassifier(n_estimators=100, random_state=50)
#model = SVC(kernel='rbf', C=1.0, gamma='scale')
#model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```
