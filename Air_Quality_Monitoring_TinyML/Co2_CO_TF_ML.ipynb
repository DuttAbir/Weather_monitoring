```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
#from sklearn.ensemble import RandomForestClassifier
#from sklearn.svm import SVC
#from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/feeds.csv')
df.head()
```

```python
df.insert(3,'Air_Quality',0)
df.head()
```

```python
co = df['CO_PPM']
co2 = df['Co2_PPM']
```

```python
for i in range(len(co2)):
  if (co[i] < 1.5 or co2[i] < 450):
    df.loc[i,'Air_Quality']= 0
  if (co[i] > 3 and co[i] < 5) or (co2[i] > 600 and co2[i] < 800):
    df.loc[i,'Air_Quality'] = 1
  if  (co[i] > 5 and co[i] < 8) or (co2[i] > 800 and co2[i] < 1200):
    df.loc[i,'Air_Quality'] = 2
  if  (co[i] > 8 and co[i] < 15) or (co2[i] > 1200 and co2[i] < 1500):
    df.loc[i,'Air_Quality'] = 3
  if (co[i] > 15):
    df.loc[i,'Air_Quality'] = 4
```

```python
df.head(100)
```

```python
# @title CO_PPM

from matplotlib import pyplot as plt
df['CO_PPM'].plot(kind='line', figsize=(8, 4), title='CO_PPM')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
# @title Co2_PPM

from matplotlib import pyplot as plt
df['Co2_PPM'].plot(kind='line', figsize=(8, 4), title='Co2_PPM')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
X = df[['CO_PPM', 'Co2_PPM']]
y = df['Air_Quality']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
```

```python
model = LogisticRegression(multi_class='multinomial', solver='newton-cg', C=1.0)
#model = RandomForestClassifier(n_estimators=100, random_state=50)
#model = SVC(kernel='rbf', C=1.0, gamma='scale')
#model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
model.coef_
```

```python
np.hstack(model.intercept_)
```

 Scatter Plot of CO and CO2 with Air Quality:

Reasoning: Visualizing the relationship between CO, CO2, and the predicted air quality can provide insights into how the model is making decisions.

How to Interpret: Different colors represent different air quality levels. You can observe how the model is classifying data points based on their CO and CO2 values.

```python
import matplotlib.pyplot as plt

# Assuming you have your dataframe 'df' with CO_PPM, Co2_PPM, and Air_Quality columns

plt.figure(figsize=(8, 6))
plt.scatter(df['CO_PPM'], df['Co2_PPM'], c=df['Air_Quality'], cmap='viridis')
plt.xlabel("CO (PPM)")
plt.ylabel("CO2 (PPM)")
plt.title("CO vs CO2 with Air Quality")
plt.colorbar(label="Air Quality")
plt.show()
```

Confusion Matrix:

Reasoning: A confusion matrix provides a visual representation of the model's performance by showing the counts of true positive, true negative, false positive, and false negative predictions. It helps in understanding the types of errors made by the model.

How to Interpret: The diagonal elements represent the correct predictions, while the off-diagonal elements represent the incorrect predictions.

```python
cm = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix using seaborn heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=[0,1,2,3,4], yticklabels=[0,1,2,3,4])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

 Decision Boundary Visualization (for Logistic Regression):

Reasoning: If you are using Logistic Regression, you can visualize the decision boundary that separates different air quality levels.

The different colored regions represent the predicted air quality levels for different combinations of CO and CO2 values. The boundary lines show where the model transitions between different predictions.

```python
x_min, x_max = X_train['CO_PPM'].min() - 1, X_train['CO_PPM'].max() + 1
y_min, y_max = X_train['Co2_PPM'].min() - 1, X_train['Co2_PPM'].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

# Predict air quality for each point in the meshgrid
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundary
plt.contourf(xx, yy, Z, alpha=0.8, cmap='viridis')
plt.scatter(X_train['CO_PPM'], X_train['Co2_PPM'], c=y_train, cmap='viridis', edgecolors='k')
plt.xlabel("CO (PPM)")
plt.ylabel("CO2 (PPM)")
plt.title("Decision Boundary of Logistic Regression")
plt.show()
```

Feature Importance Plot:

Reasoning: This plot shows the relative importance of each feature in the model's prediction process. It helps identify which features have the most influence on the model's decisions. This can be particularly useful for understanding the model's behavior and potentially for feature selection or engineering.
How to interpret: Features with higher importance scores have a greater impact on the model's predictions. You can use this information to focus on the most important features for further analysis or to potentially remove less important features.

```python
importances = model.coef_[0]  # Get feature importance coefficients
features = ['CO_PPM', 'Co2_PPM']  # Replace with your feature names

plt.figure(figsize=(8, 6))
plt.bar(features, importances)
plt.xlabel("Features")
plt.ylabel("Importance")
plt.title("Feature Importance Plot")
plt.show()
```
